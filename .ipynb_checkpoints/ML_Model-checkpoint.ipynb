```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generate 4 clusters of random data.
from sklearn.datasets.samples_generator import make_blobs

data = pd.read_csv("results.csv")

data["HomeWin"] = np.where(data['home_score']>data['away_score'], 1, 0)
data["HomeCity"] = np.where(data['home_team']>data['country'], 1, 0)
data["Tie"] = np.where(data['HomeWin'] + data)

data.head(50)
```

```python
# Plot the data
plt.scatter(data["HomeWin"], data["away_score"])
```

```python
# Use n_clusters=4 as the k value
# We can see from the plot above that there are 4 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
data = data[["HomeWin", "away_score"]]
```

```python
# Fit the model to the data
kmeans.fit(data)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(data)
```

```python
# Plot the predicted clusters to see if the model predicted the correct clusters
# This is visual validation that the model was trained correctly.
plt.scatter(data["HomeWin"], data["away_score"], c=predicted_clusters, s=50, cmap='viridis')
```
